# Monitoring & Evaluation in Global Health: A Connected Process

Monitoring and Evaluation (M&E) are essential components in global health projects, ensuring that goals are met, and real-time adjustments are made based on evidence. The success of M&E lies in how well it is connected to the broader project implementation process and the realities on the ground. Here's an example illustrating the connected process of M&E using a maternal and child health project.

---

## Example: M&E in a Child Nutrition Project in Ethiopia

Let’s say you are the M&E lead for a child nutrition project in Ethiopia aimed at reducing malnutrition rates in rural areas by improving access to nutritious foods and educating communities about balanced diets. The project is funded by an international donor, and the proposal outlines ambitious targets for reducing malnutrition rates by 30% over three years.

### 1. **Understanding the Local Context**
Upon arrival, you engage with local partners, including community health workers and local government officials. You quickly realize that although the project proposal includes distributing fortified foods, local communities have deep-rooted cultural beliefs around food consumption. These beliefs must be considered for the project to be effective.

$$\textbf{Action}:$$ You postpone developing the M&E framework until you've conducted meetings and field visits. You listen to community members and health workers, ensuring that the project goals align with local practices and beliefs. This way, you create a plan that integrates cultural aspects into both the project implementation and monitoring process.

### 2. **Collaboration for Tool Development**
Instead of designing tools in isolation, you work directly with local health workers to co-create simple, effective M&E tools. Together, you develop a nutrition assessment tool that is culturally appropriate and easy for community health workers to use in tracking children's growth and diet.

$$\textbf{Action}:$$ The checklist and scorecards developed not only measure malnutrition rates but also track community engagement in educational sessions and the number of households adopting the new dietary practices. This connected process ensures that both the intervention and the M&E strategy are in sync, reflecting the realities on the ground.

### 3. **Data Integration with Project Implementation**
As the project rolls out, the local health workers use the assessment tools during routine visits to households. The data collected (on child nutrition status and family engagement) is integrated into the project's decision-making process. For example, when data shows that one village has slower adoption of nutrition practices, the project team increases support and provides more educational sessions in that area.

$$\textbf{Action}:$$ You ensure that the M&E process is dynamic, regularly feeding data back into the project to inform adjustments. The data collection isn’t just for reporting to the donor; it becomes an active part of how the project is implemented, creating a feedback loop where both the intervention and M&E processes continuously inform each other.

### 4. **Iterative Process and Real-time Adjustments**
Halfway through the project, your data shows that while malnutrition rates are decreasing, they’re not dropping as quickly as expected. Upon further investigation, you discover that many households face food shortages during the dry season, which limits the impact of your nutrition interventions.

$$\textbf{Action}:$$ Based on this finding, the project team modifies the intervention by adding a food voucher component during the lean season. You adjust the M&E framework to track the new intervention, ensuring that the project remains responsive to emerging challenges. This shows how connected M&E enables the project to adapt and address issues in real time.

### 5. **Avoiding a Pre-Formed Plan**
Instead of coming to Ethiopia with a rigid M&E plan developed based solely on the project proposal, you take a step back and involve local stakeholders. By asking questions and observing their capacity, needs, and local practices, you develop a flexible and adaptive M&E system that reflects the ground realities rather than sticking to preconceived notions.

---

## Key Elements of a Connected M&E Process:
- **Listening and Learning First**: Ensure the M&E plan is grounded in the local context by asking questions and understanding the needs of the community.
- **Co-Creation**: Work alongside local teams to develop practical and culturally appropriate tools that align with both the intervention and the M&E process.
- **Feedback Loops**: Continuously feed M&E data back into the project to inform real-time decision-making and adjustments.
- **Flexibility**: Allow the M&E process to evolve in response to challenges and emerging needs, ensuring it stays relevant throughout the project.

In this example, the connected process ensures that M&E is not a separate or static activity, but a dynamic part of the project’s success. By integrating M&E with the intervention and remaining flexible, you create a system that improves both project implementation and its outcomes, leading to better health impacts.


# Example: Analyzing and Visualizing Malnutrition Data in a Child Nutrition Project

## Step 1: Simulating Data
Let's simulate data for two regions in Ethiopia, showing the malnutrition rates before and after the intervention.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Simulate data for two regions
data = {
    'Region': ['Region A', 'Region A', 'Region B', 'Region B'],
    'Time': ['Before Intervention', 'After Intervention', 'Before Intervention', 'After Intervention'],
    'Malnutrition Rate (%)': [40, 25, 35, 20],
    'Population': [1000, 1000, 800, 800]  # Total population in each region
}

df = pd.DataFrame(data)

# Calculate the number of malnourished children
df['Malnourished Children'] = (df['Malnutrition Rate (%)'] / 100) * df['Population']
df
```

# Step 2: Visualization
Now we will create a bar plot to visualize the change in malnutrition rates before and after the intervention for both regions.

```python
# Create the bar plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the data
before = df[df['Time'] == 'Before Intervention']
after = df[df['Time'] == 'After Intervention']

# Bar positions
x = np.arange(len(before['Region']))

# Bar widths
bar_width = 0.35

# Plot before and after intervention data
bars_before = ax.bar(x - bar_width/2, before['Malnutrition Rate (%)'], bar_width, label='Before Intervention', color='lightcoral')
bars_after = ax.bar(x + bar_width/2, after['Malnutrition Rate (%)'], bar_width, label='After Intervention', color='lightgreen')

# Adding annotations for numbers and percentages
for i, rect in enumerate(bars_before):
    height = rect.get_height()
    malnourished_before = before.iloc[i]['Malnourished Children']
    ax.annotate(f'{int(malnourished_before)} children\n({height}%)',
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

for i, rect in enumerate(bars_after):
    height = rect.get_height()
    malnourished_after = after.iloc[i]['Malnourished Children']
    reduction_percentage = before.iloc[i]['Malnutrition Rate (%)'] - height
    ax.annotate(f'{int(malnourished_after)} children\n({height}%)\n↓{reduction_percentage}%',
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

# Adding labels and title
ax.set_xlabel('Regions')
ax.set_ylabel('Malnutrition Rate (%)')
ax.set_title('Malnutrition Rates Before and After Intervention')
ax.set_xticks(x)
ax.set_xticklabels(before['Region'])
ax.legend()

# Display plot
plt.tight_layout()
plt.show()
```
