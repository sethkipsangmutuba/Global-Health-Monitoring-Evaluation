# Monitoring & Evaluation in Global Health: A Disconnected Process

## Example Overview
In global health, a major challenge arises when there is a disconnect between the realities on the ground and the designs created by those working remotely or with limited contextual knowledge. This disconnection often happens in the phases of proposal writing, project implementation, and report writing.

## Case Example: HIV Care and Treatment Project in Côte d'Ivoire
A few years ago, I served as a consultant, tasked with writing a proposal for an HIV care and treatment project in Côte d'Ivoire. While I had significant experience in monitoring and evaluation (M&E), particularly in HIV-related projects, and a deep understanding of the U.S. government’s PEPFAR program (the funding organization), I faced challenges because I had never been to Côte d'Ivoire. My role was to provide technical expertise in monitoring and evaluation to ensure the proposal met the donor's standards.

## Key Challenge: Disconnection Between Proposal Design and Local Realities
Despite being well-versed in the technical requirements and donor expectations, I was disconnected from the specific context of Côte d'Ivoire. My focus was to create a technically sound M&E section that would appeal to reviewers, but I relied on local staff and country-specific experts to ensure the project made sense for the Côte d'Ivoire context. I had minimal direct interaction with those on the ground during this phase.

## Steps Taken to Mitigate the Disconnection
1. **Consultation with Local Experts**: Though I wasn’t based in Côte d'Ivoire, I consulted local staff who had knowledge of the realities in the country. Their insights were invaluable in shaping certain aspects of the proposal.

2. **Highlighting Gaps in Knowledge**: In my work, I was clear to note areas where I lacked country-specific knowledge and emphasized that additional adjustments would be necessary during implementation to ensure the project was aligned with local needs.

3. **Trust in Local Implementation**: While I played a key role in the proposal stage, I trusted that the organization’s local office, staffed by people with in-depth knowledge of the context, would tailor the project during implementation to better fit the country’s realities.

## Lessons Learned
This example underscores the importance of **contextual knowledge** in designing effective M&E frameworks for global health projects. While technical expertise in M&E practices is essential, successful project implementation often requires local insights. The disconnection in this process can lead to challenges not just in the proposal phase, but also during project launch and report writing when data is interpreted without a nuanced understanding of the local environment.

This example highlights a common challenge in global health M&E, where **coordination between technical consultants and local experts** is crucial to bridge the gap between **funding requirements** and **on-the-ground realities**.


# Disconnected Process: Example in Monitoring & Evaluation

In a disconnected process, the design of a health project (or its monitoring and evaluation) may not be fully aligned with the realities on the ground. Let's simulate such a scenario in Python, where a health intervention is designed without fully understanding the local context, leading to a mismatch in expected outcomes.

# Use Case:
We have 10 health facilities, and a proposal was made assuming all facilities would show similar improvements after an intervention. However, the intervention’s actual implementation results differ because the facilities' initial conditions were not considered in the planning. We'll:

Simulate data for expected and actual outcomes of the intervention.
Analyze the mismatch between expectations and reality.
Visualize the results with annotated percentages showing the disconnection.
Python Example: Disconnected Process in Health Facilities

# Step 1: Simulate Data for Expected vs. Actual Outcomes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulate data for 10 health facilities
np.random.seed(42)  # For reproducibility
facilities = ['Facility A', 'Facility B', 'Facility C', 'Facility D', 'Facility E',
              'Facility F', 'Facility G', 'Facility H', 'Facility I', 'Facility J']

# Expected outcomes after intervention (in percentages)
expected_outcome = np.random.randint(70, 90, size=10)  # Expected success rate

# Actual outcomes after intervention (in percentages)
# Actual results are lower due to contextual issues not considered in the design
actual_outcome = expected_outcome - np.random.randint(10, 30, size=10)

# Create a DataFrame
data = pd.DataFrame({
    'Facility': facilities,
    'Expected': expected_outcome,
    'Actual': actual_outcome
})

# Display the data
data
```

# Step 2: Analyze the Mismatch
We calculate the percentage gap (or disconnection) between the expected and actual outcomes for each facility.

```python
# Calculate the disconnection percentage (gap between expected and actual)
data['Disconnection (%)'] = ((data['Expected'] - data['Actual']) / data['Expected']) * 100

# Calculate average disconnection
average_disconnection = data['Disconnection (%)'].mean()

# Display the updated data
data
```

# Step 3: Visualize the Disconnected Process
We'll create a bar chart comparing the expected and actual outcomes for each facility and annotate the disconnection percentages.

```python
# Plot the expected and actual outcomes
fig, ax = plt.subplots(figsize=(10, 6))

# Plot bars
bar_width = 0.35
index = np.arange(len(facilities))

# Bars for 'Expected' and 'Actual'
bars_expected = ax.bar(index, data['Expected'], bar_width, label='Expected Outcome', color='lightgreen')
bars_actual = ax.bar(index + bar_width, data['Actual'], bar_width, label='Actual Outcome', color='lightcoral')

# Add text annotations with disconnection percentages
for i, v in enumerate(data['Disconnection (%)']):
    ax.text(i + bar_width / 2, data['Actual'][i] + 1, f'{v:.1f}%', ha='center', color='black')

# Labels and title
ax.set_xlabel('Health Facility', fontsize=12)
ax.set_ylabel('Compliance Rate (%)', fontsize=12)
ax.set_title('Expected vs. Actual Outcomes: Disconnected Process in Health Facilities', fontsize=14)
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(facilities)
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
